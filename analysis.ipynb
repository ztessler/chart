# Time-averaged model

```python
target = '50_Percentile_TSM (g/m3)'
```

```python
Xy = pd.merge(left=tsm[['Basin', 'Month Composite', target]], right=data, on=['Basin', 'Month Composite'], how='left')
display(Xy.head())
print(Xy.shape)
```

```python
Xy.drop('Month Composite', axis=1).groupby('Basin').mean()
```

```python
np.unique(Xy['Basin'])
```

```python
y = Xy[['Basin',target]].groupby('Basin').mean()
y.head()
```

```python
X = Xy.drop(['Month Composite', target], axis=1).groupby('Basin').mean()
print(X.columns)
X = X[['RiverDischarge', 'ReservoirRelease', 'SedimentFlux', 'PrecipitationZonalMean']]
X.head()
```

```python
X.shape, y.shape
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
ypred = lr.predict(X)
```

```python
residuals = y - ypred
```
